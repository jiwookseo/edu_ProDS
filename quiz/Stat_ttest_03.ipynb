# 다음 문제를 풀이하시오

### Q1. "bike.csv" 파일을 Pandas 라이브러리를 활용하여 불러온 후 df 객체에 저장하시오.

```python
import pandas as pd
df = pd.read_csv("../dataset/bike.csv")
```

### Q2. df객체의 "datetime"변수에서 월 정보를 뽑아 df객체에 "month" 변수를 새로 만드시오.

```python
df["month"] = df.datetime.astype("datetime64").dt.month
df.tail()
```

### Q3. 6월과 12월의 자전거 대여 차이를 알아보고자 한다. "registered" 변수와 독립 2표본 t검정을 사용하여 결과를 확인하시오.

```python
from scipy.stats import ttest_ind
ttest_ind(df.loc[df.month == 6, "registered"], df.loc[df.month == 12, "registered"])
```

### Q4. df객체의 "datetime"변수에서 요일 정보를 뽑아 df객체에 "wday" 변수를 새로 만드시오

```python
df["wday"] = df.datetime.astype("datetime64").dt.dayofweek # 월~일: 0~6
```

### Q5. 주말(토, 일)이 1이고 나머지 요일이 0인 "wend" 변수를 df객체에 만드시오.

```python
df["wend"] = (df.wday >= 5) + 0
pd.crosstab(df.wday, df.wend)
# (pd.Series(["2021-04-29"]).astype("datetime64").dt.dayofweek >= 5) + 0 # 월~일: 0~6
```

### Q6. 주중과 주말에 따른 "casual" 변수값의 차이가 있는지 독립 2표본 t검정을 사용하여 결과를 확인하시오.

```python
ttest_ind(df.loc[df.wend==1, "casual"], df.loc[df.wend==0, "casual"])
```
