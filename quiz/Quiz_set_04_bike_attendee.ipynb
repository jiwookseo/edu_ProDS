### Q5. pandas.DataFrame의 join() 을 활용하여 weather과 rental 데이터를 병합하시오.
(병합한 데이터는 df_join에 저장하시오)
 * pandas.to_timedelta() 를 활용하시오
 * pandas.Series.astype() 를 활용하시오
 
※ 힌트: 일 데이터에 to_timedelta() 함수로 변환된 시간 증분을 더하여 두 객체의 시간 형식을 맞추시오

```python
# weather["datetime"] = weather.date.astype("dateime64[ns]") + pd.to_timedelta(weather.hour, "h")
weather["datetime"] = weather.date + pd.to_timedelta(weather.hour, "h").astype("str").apply(lambda x: " " + x.split()[2])
weather = weather.set_index("datetime")
weather
```

```python
# rental.datetime = rental.datetime.astype("datetime64[ns]")
# rental.datetime = pd.to_datetime(rental.datetime)
rental = rental.set_index("datetime")
rental
```

```python
df_join = weather.join(rental)
df_join
```

### Q6. df_join의 변수별 결측치를 확인하시오.

```python
df_join.isna().sum().reset_index()
```
