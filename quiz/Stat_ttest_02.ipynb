### Q1. "class_scores.csv" 파일을 Pandas 라이브러리를 활용하여 불러온 후 df 객체에 저장하시오.

```python
import pandas as pd
df = pd.read_csv("../dataset/class_scores.csv")
```

### Q2. 성별에 따른 수학 점수의 차이를 확인하기 위해 독립 2표본 t검정을 실시하고 그 결과를 확인하시오.

```python
math_M = df.loc[df.gender == "M", "Math"]
math_F = df.loc[df.gender == "F", "Math"]
```

```python
from scipy.stats import ttest_ind
stat, p = ttest_ind(math_F, math_M)
stat, p
```

### Q3. 1학년과 3학년의 영어 점수의 차이를 확인하기 위해 독립 2표본 t검정을 실시하고 그 결과를 확인하시오.

```python
eng_1 = df.loc[df.grade == 1, "English"]
eng_3 = df.loc[df.grade == 3, "English"]
stat, p = ttest_ind(eng_1, eng_3)
stat, p
```

### Q4. 2학년과 3학년 중 평균점수가 50점 이상인 학생들만 따로 뽑아서 과학 점수의 차이를 확인해보려 한다. 독립 2표본 t검정을 실시하고 그 결과를 확인하시오.

```python
df
```

```python
df["mean"] = df.iloc[:, 4:].mean(axis=1)
scn_2 = df.loc[(df.grade == 2) & (df["mean"] >= 50), "Science"]
scn_3 = df.loc[(df.grade == 3) & (df["mean"] >= 50), "Science"]
stat, p = ttest_ind(scn_2, scn_3)
stat, p
```
