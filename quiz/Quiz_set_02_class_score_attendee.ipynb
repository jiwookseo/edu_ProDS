### Q9. 2학년의 학번, 학년, 성적 전체를 추출하여 df_sub 객체에 저장하시오.

```python
df_sub = df.loc[df.grade == 2, ["Stu_ID", "grade", *df.columns[-5:]]]
df_sub
```

### Q10. df_sub를 기준으로 학생 개개인의 성적 평균을 계산하고 학번과 같이 "score_avg.csv"로 저장하시오.

```python
mean_ = df_sub.iloc[:, 2:].mean(axis=1)
score_avg = pd.DataFrame({"Stu_ID": df_sub.Stu_ID, "mean": mean_})
score_avg.to_csv("score_avg.csv", index=False)
score_avg
```
